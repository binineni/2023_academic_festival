```python
from sklearn.datasets import make_classification
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
datapath = '/content/drive/MyDrive/Colab Notebooks/경경대 학술제(?)/data'
```

```python
np.random.choice(range(0,2))
```

```python
#독립변수 개수별로 / informative한 독립변수 개수별로 데이터셋 저장장
X_2_1, y_2_1 = make_classification(n_samples=10000, n_features=2, n_informative=1, n_redundant=0 ,flip_y=0.05 ,n_clusters_per_class=1, random_state=42)
X_2_2, y_2_2 = make_classification(n_samples=10000, n_features=2, n_informative=2, n_redundant=0 ,flip_y=0.05 ,n_clusters_per_class=1, random_state=42)

X_5_1, y_5_1 = make_classification(n_samples=10000, n_features=5, n_informative=1,n_redundant=0 ,flip_y=0.05 ,n_clusters_per_class=1, random_state=42)
X_5_3, y_5_3 = make_classification(n_samples=10000, n_features=5, n_informative=3,n_redundant=1 ,flip_y=0.05 ,n_clusters_per_class=1, random_state=42)
X_5_5, y_5_5 = make_classification(n_samples=10000, n_features=5, n_informative=5,n_redundant=0 ,flip_y=0.05 ,n_clusters_per_class=1, random_state=42)

X_10_1, y_10_1 = make_classification(n_samples=10000, n_features=10, n_informative=1,n_redundant=0 ,flip_y=0.05,n_clusters_per_class=1, random_state=42)
X_10_3, y_10_3 = make_classification(n_samples=10000, n_features=10, n_informative=3,n_redundant=0 ,flip_y=0.05,n_clusters_per_class=1, random_state=42)
X_10_5, y_10_5 = make_classification(n_samples=10000, n_features=10, n_informative=5,n_redundant=0 ,flip_y=0.05,n_clusters_per_class=1, random_state=42)
X_10_7, y_10_7 = make_classification(n_samples=10000, n_features=10, n_informative=7,n_redundant=0 ,flip_y=0.05,n_clusters_per_class=1, random_state=42)
X_10_9, y_10_9 = make_classification(n_samples=10000, n_features=10, n_informative=9,n_redundant=0 ,flip_y=0.05,n_clusters_per_class=1, random_state=42)

print(X_2_1[:5]) # X 데이터의 첫 5개 행 출력
print(y_2_2[:5]) # y 데이터의 첫 5개 행 출력
```

```python
# X와 y 데이터를 하나의 데이터프레임으로 합치기
df_1 = pd.DataFrame(data=X_2_1, columns=['X1', 'X2'])
df_1['y'] = y_2_1
df_2 = pd.DataFrame(data=X_2_2, columns=['X1', 'X2'])
df_2['y'] = y_2_2

# CSV 파일로 저장
df_1.to_csv(datapath + '/bin_X2_1_data.csv', index=False)
df_2.to_csv(datapath + '/bin_X2_2_data.csv', index=False)

# X와 y 데이터를 하나의 데이터프레임으로 합치기
df_3 = pd.DataFrame(data=X_5_1, columns=['X1', 'X2','X3','X4','X5'])
df_3['y'] = y_5_1
df_4 = pd.DataFrame(data=X_5_3, columns=['X1', 'X2','X3','X4','X5'])
df_4['y'] = y_5_3
df_5 = pd.DataFrame(data=X_5_5, columns=['X1', 'X2','X3','X4','X5'])
df_5['y'] = y_5_5

# CSV 파일로 저장
df_3.to_csv(datapath + '/bin_X5_1_data.csv', index=False)
df_4.to_csv(datapath + '/bin_X5_3_data.csv', index=False)
df_5.to_csv(datapath + '/bin_X5_5_data.csv', index=False)

# X와 y 데이터를 하나의 데이터프레임으로 합치기
df_6 = pd.DataFrame(data=X_10_1, columns=['X1', 'X2','X3','X4','X5','X6','X7','X8','X9','X10'])
df_6['y'] = y_10_1
df_7 = pd.DataFrame(data=X_10_3, columns=['X1', 'X2','X3','X4','X5','X6','X7','X8','X9','X10'])
df_7['y'] = y_10_3
df_8 = pd.DataFrame(data=X_10_5, columns=['X1', 'X2','X3','X4','X5','X6','X7','X8','X9','X10'])
df_8['y'] = y_10_5
df_9 = pd.DataFrame(data=X_10_7, columns=['X1', 'X2','X3','X4','X5','X6','X7','X8','X9','X10'])
df_9['y'] = y_10_7
df_10 = pd.DataFrame(data=X_10_9, columns=['X1', 'X2','X3','X4','X5','X6','X7','X8','X9','X10'])
df_10['y'] = y_10_9

# CSV 파일로 저장
df_6.to_csv(datapath + '/bin_X10_1_data.csv', index=False)
df_7.to_csv(datapath + '/bin_X10_3_data.csv', index=False)
df_8.to_csv(datapath + '/bin_X10_5_data.csv', index=False)
df_9.to_csv(datapath + '/bin_X10_7_data.csv', index=False)
df_10.to_csv(datapath + '/bin_X10_9_data.csv', index=False)
```
